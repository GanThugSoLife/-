```python
import numpy as np
import CoolProp.CoolProp as ct
import matplotlib.pyplot as plt
T=273.15  #температура в K
p=101325  #давление 1 атм
V=64      #объем м3
R=8.3145   #Универсальная газовая постоянная
M=0.016   #Моллерная масса метана
Fluid_name="CH4"
T=np.linspace(100,150,1000) #Задание температуры
m1=(p*V*M)/R*T #Масса по школьному 
D=ct.PropsSI("D","T",T,"P",p,"CH4") #плотность по CoolProp
m2=D*V  #Масса через Coolprop
plt.plot(T,m1,"b")
plt.plot(T,m2,"r")
plt.xlabel("T")
plt.ylabel("m")
plt.title("Зависимость m от T при постоянном давлении")
plt.legend(['m школьная','m Coolprop'])
```
